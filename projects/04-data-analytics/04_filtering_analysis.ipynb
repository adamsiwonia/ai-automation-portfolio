```python
df[(df["total_bill"] > 20) & (df["tip"] > 3)]
```

```python
df[(df["day"] == "Sun") | (df["day"] == "Sat") ]
```

Operator & i | w pandas

& → logiczne AND element po elemencie

| → logiczne OR element po elemencie

Reguła do zapamiętania

W pandas:

każdy warunek → nawias

& = i

| = lub

```python
df[df["day"].isin(["Sat", "Sun"])]
```

isin() to metoda biblioteki pandas, używana do sprawdzania, czy wartości znajdują się w określonym zbiorze wartości. Najczęściej stosuje się ją do filtrowania DataFrame lub Series na podstawie listy dopuszczalnych elementów.

```python
df[df["time"] == "Dinner"]
```

W pandas operator == jest operatorem porównania równości, używanym do wektorowego porównywania danych w obiektach takich jak Series i DataFrame.


Zasada praktyczna

Jeśli myślisz „czy to jest jedna z tych wartości” → użyj isin()
Jeśli myślisz „czy to jest dokładnie to” → użyj ==
