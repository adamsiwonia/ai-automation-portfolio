1. Dlaczego każdy warunek musi być w nawiasach?

Powody są dwa:
kolejność wykonywania operatorów oraz sposób działania pandas Series.

a) Kolejność operatorów (operator precedence)

Python najpierw wykonuje operatory porównań (>, <, ==), a dopiero potem operatory logiczne (&, |).

Czyli:

df["total_bill"] > 20


zwraca Series wartości True/False.

Dopiero dwie takie Series można połączyć operatorem &.

b) & działa na CAŁYCH Series, nie na pojedynczych wartościach

W pandas:

df["total_bill"] > 20 → Series bool

df["tip"] > 3 → Series bool

& łączy element po elemencie (wektorowo)

To działa tylko wtedy, gdy & dostanie dwie gotowe Series.

2. Co by się stało bez nawiasów?

Załóżmy, że napisałbyś:

df[df["total_bill"] > 20 & df["tip"] > 3]


Python przeczyta to w takiej kolejności:

df["total_bill"] > (20 & df["tip"]) > 3


I tu zaczyna się problem.

Co jest nie tak?

20 & df["tip"]
→ Python próbuje wykonać operację bitową &
→ między liczbą 20 a Series floatów

To kończy się błędem w stylu:

TypeError: unsupported operand type(s) for &: 'int' and 'Series'


Albo:

ValueError: The truth value of a Series is ambiguous

3. Dlaczego and / or też NIE działają?

Częsty błąd początkujących:

df[(df["total_bill"] > 20) and (df["tip"] > 3)]


To zawsze jest błąd, ponieważ:

and / or działają tylko na pojedynczych wartościach True/False

pandas Series zawiera wiele wartości

Python nie wie, czy Series jest „prawdziwa czy fałszywa jako całość”.

1. df[...] — skrótowy operator indeksowania

df[...] to uniwersalny, przeciążony operator, który zachowuje się różnie w zależności od tego, co wstawisz do środka.

Najczęstsze znaczenia:
df["total_bill"]              # pojedyncza kolumna (Series)
df[["total_bill", "tip"]]     # wiele kolumn (DataFrame)
df[mask]                      # filtrowanie wierszy (gdzie mask = True/False)


Przykład:

df[df["total_bill"] > 25]


➡️ wybiera wiersze, ale nie kolumny jednocześnie
➡️ kolumny trzeba wybrać w osobnym kroku

df[df["total_bill"] > 25][["total_bill", "tip", "day"]]

Wady df[...]

brak jawnego rozdzielenia: wiersze vs kolumny

łatwiej o:

nieczytelny kod

ostrzeżenie SettingWithCopyWarning

trudniej zrozumieć kod po kilku tygodniach

2. df.loc[...] — indeksowanie etykietami (jawne i bezpieczne)

loc zawsze działa według schematu:

df.loc[wiersze, kolumny]


Twoje wyrażenie:

df.loc[df["total_bill"] > 25, ["total_bill", "tip", "day"]]


czytamy dosłownie:

„Daj wiersze, gdzie total_bill > 25, oraz kolumny: total_bill, tip, day”

Zalety loc

jednoznaczność

większa czytelność

brak niejasnych zachowań

bezpieczniejsze przy modyfikacji danych

3. Kiedy loc jest wyraźnie czytelniejsze?
✅ Gdy wybierasz wiersze i kolumny jednocześnie
df.loc[df["day"] == "Sun", ["total_bill", "tip"]]


vs

df[df["day"] == "Sun"][["total_bill", "tip"]]  # mniej czytelne

✅ Gdy masz więcej warunków
df.loc[
    (df["day"] == "Sun") & (df["total_bill"] > 25),
    ["total_bill", "tip", "size"]
]


To jest:

logiczne

czytelne

łatwe do debugowania

✅ Gdy zmieniasz dane
df.loc[df["total_bill"] > 25, "tip"] = 5


❌ Unikaj:

df[df["total_bill"] > 25]["tip"] = 5


To częsty błąd początkujących (kopie vs widoki).

4. Reguła praktyczna (bardzo ważna)


df[...] → szybki skrót do odczytu

df.loc[...] → precyzyjna selekcja i edycja danych

Jeśli:

kod ma być czytelny

wrócisz do niego za miesiąc

planujesz modyfikacje danych

➡️ używaj loc

df[...] zostawia to do domyślenia.
