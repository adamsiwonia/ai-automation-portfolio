aggregation

```python
df.groupby("day")["total_bill"].agg(["mean", "max", "min", "count"])
```

group by more than one column

```python
df.groupby(["day","sex"])["total_bill"].mean()
```

groupby with filter

```python
df[df["total_bill"] >30].groupby("day")["tip"].mean()
```

Conclusions 
 Profitability by Day

Saturdays and Sundays generate the highest average bills, indicating stronger customer spending on weekends.

Average tips are also highest on weekends, which suggests higher overall profitability and better tipping behavior.

Weekdays (especially Thursday) show lower average bills and tips, making them less profitable in comparison.

 Profitability by Time of Day

Dinner is significantly more profitable than Lunch.

The average bill during Dinner is much higher, indicating customers spend more in the evening.

Lunch appears to be more volume-based, while Dinner is value-based.

 Final Insight

The most profitable period is weekend dinners, especially Saturdays.
Businesses should prioritize staffing, promotions, and premium offerings during this time to maximize revenue.
